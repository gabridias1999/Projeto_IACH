```python
import pandas as pd
import numpy as np
```

```python
games_df = pd.read_csv('steam.csv')
ratings_df = pd.read_csv('final_dataset1.csv')
```

```python
games_df.head(n=10)
```

```python
games_dfgenres = games_df.iloc[:,9:10]
games_dfgenres.head(n=35)
```

```python
games_df['genres'] = games_df.genres.str.split(';')
games_df.head()
```

```python
pd.set_option('max_columns', None)

#Copiar o game dataframe para um novo 

gamesWithGenres_df = games_df.copy()

#Para cada linha no dataframe, da lista de gêneros e colocar um 1 na coluna correspondente

for index, row in games_df.iterrows():
    for genre in row['genres']:
        gamesWithGenres_df.at[index, genre] = 1
        
#Preencher os valores NaN com 0 para mostrar que um filme não tem o gênero daquela coluna
gamesWithGenres_df = gamesWithGenres_df.fillna(0)
gamesWithGenres_df.head()
```

```python
ratings_df.head()
```

```python
userInput = [
            {'name':'Counter-Strike', 'rating':1},
            {'name':'Earth 2160', 'rating':4.5},
            {'name':'Depths of Peril', 'rating':2},
            {'name':"Portal 2", 'rating':5},
            {'name':"X Rebirth", 'rating':4.5}
         ] 
inputGames = pd.DataFrame(userInput)
inputGames
```

```python
#Filtrar os jogos por nome
inputId = games_df[games_df['name'].isin(inputGames['name'].tolist())]

#juntar as tabelas, por nome do jogo para obter o appid.
inputGames = pd.merge(inputId, inputGames)

inputGames = inputGames.loc[:, ['appid','name','rating']]

inputGames
```

```python
#Filtar os jogos do input
userGames = gamesWithGenres_df[gamesWithGenres_df['appid'].isin(inputGames['appid'].tolist())]
userGames


```

```python

userGames = userGames.reset_index(drop=True)

#Dropar as colunas que não nos interessam

userGenreTable = userGames.drop(columns=['release_date', 'english','developer','appid','name','publisher','platforms','required_age','categories','genres','steamspy_tags','achievements','positive_ratings','negative_ratings','average_playtime','median_playtime','owners','price'])
userGenreTable
```

```python
inputGames['rating']
```

```python
#Calcular os pesos das categorias dado os ratings atribuidos
userProfile = userGenreTable.transpose().dot(inputGames['rating'])

userProfile
```

```python
#Obter as categorias dos jogos no dataframe original 
genreTable = gamesWithGenres_df.set_index(gamesWithGenres_df['appid'])

#Dropar novamente o que não necessitamos
genreTable = genreTable.drop(columns=['release_date', 'english','developer','appid','name','publisher','platforms','required_age','categories','genres','steamspy_tags','achievements','positive_ratings','negative_ratings','average_playtime','median_playtime','owners','price'])
genreTable.head(n=30)

```

```python
genreTable.shape
```

```python
#
recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df.head(n=30)
```

```python
#Ordenar por ordem decrescente as Recomendações para obter os mais próximos do perfil do user
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)

recommendationTable_df.head()
```
